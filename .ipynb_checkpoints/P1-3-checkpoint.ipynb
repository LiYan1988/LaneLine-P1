```python
rate_list = []

def detect_lane_pipeline(img):
    """
    Detects lanes
    """
    # Resize image
    if img.shape[0]!=960:
        img = cv2.resize(img, (960, 540))
    
    # Convert to gray scale
    img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    
    # Gaussian blur
    gaussian_kernel = 5
    img_gray = cv2.GaussianBlur(img_gray, (gaussian_kernel, gaussian_kernel), 0)
    
    # Canny
    canny_th_low = 10
    canny_th_high = 200
    img_gray = cv2.Canny(img_gray, canny_th_low, canny_th_high)
    
    # Mask
    img_gray_mask = np.zeros_like(img_gray)
    cv2.fillPoly(img_gray_mask, [vertices], 255)
    img_gray = cv2.bitwise_and(img_gray, img_gray_mask)
    
    img_color_mask = np.zeros_like(img)
    cv2.fillPoly(img_color_mask, [vertices], [255, 0, 0])
    img_color_masked = cv2.bitwise_and(img, img_color_mask)
    img_color_masked_intensity = img_color_masked.mean(axis=(0, 1, 2))/1.74796410436
    
    img_gray = img_gray/img_color_masked_intensity
    img_gray = img_gray.astype(np.uint8)
    
    # Hough 
    rho = 1
    theta = np.pi/180
    threshold = 30
    min_line_length = 30
    max_line_gap = 30
    lines = cv2.HoughLinesP(img_gray, rho, theta, threshold, np.array([]), 
                            min_line_length, max_line_gap)
    img_blank = np.zeros_like(img)
    
    # Detect lane
    lane_left = []
    lane_right = []
    for line in lines:
        for x1, y1, x2, y2 in line:
            if (y2-y1)/(x2-x1)>0:
                lane_right.append([x1, y1])
                lane_right.append([x2, y2])
            else:
                lane_left.append([x1, y1])
                lane_left.append([x2, y2])

    points_left = np.squeeze(np.array(lane_left))
    points_right = np.squeeze(np.array(lane_right))
    
    try:

        y_min = min(min(points_left[:, 1]), min(points_right[:, 1]))
        y_max = max(max(points_left[:, 1]), max(points_right[:, 1]))
    #     y_min = np.int(y_max*0.3 + y_min*0.7)


        fit_left = np.polyfit(points_left[:, 0], points_left[:, 1], 1)
        fit_right = np.polyfit(points_right[:, 0], points_right[:, 1], 1)

        x_left_min = np.int((y_min-fit_left[1])/fit_left[0])
        x_left_min = min(959, x_left_min)
        x_left_max = np.int((y_max-fit_left[1])/fit_left[0])
        x_left_max = min(959, x_left_max)
        x_left = [x_left_min, x_left_max]

        y_left = np.polyval(fit_left, x_left).astype(np.int)
        x_right_min = np.int((y_min-fit_right[1])/fit_right[0])
        x_right_min = min(959, x_right_min)
        x_right_max = np.int((y_max-fit_right[1])/fit_right[0])
        x_right_max = min(959, x_right_max)
        x_right = [x_right_min, x_right_max]
        y_right = np.polyval(fit_right, x_right).astype(np.int)

        cv2.line(img_blank, (x_left_min, y_min), (x_left_max, y_max), [255, 0, 0], 10)
        cv2.line(img_blank, (x_right_min, y_min), (x_right_max, y_max), [255, 0, 0], 10)

        img_gray = np.dstack((img_gray, img_gray, img_gray))
        img_combo = cv2.addWeighted(img_gray, 1, img_blank, 0.8, 0)
    except:
        img_gray = np.dstack((img_gray, img_gray, img_gray))
        img_combo = cv2.addWeighted(img_gray, 1, img_blank, 0.8, 0)
    

    return img_combo
```

```python
video_input = 'test_videos/challenge.mp4'
video_output = 'test_videos_output/challenge.mp4'
clip_in = VideoFileClip(video_input)
clip_out = clip_in.fl_image(detect_lane_pipeline) 
%time clip_out.write_videofile(video_output, audio=False)
HTML("""
<video width="960" height="540" controls>
  <source src="{0}">
</video>
""".format(video_output))
```

```python
cap = cv2.VideoCapture('test_videos/challenge.mp4')
count = 0
suc = True
img_color_masked_intensity = []
mask_max_intensity = []
while count<244:
    suc, img = cap.read()
    count += 1
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img.mean(axis=(0, 1, 2))
    img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

    # Mask
    img_gray_mask = np.zeros_like(img_gray)
    cv2.fillPoly(img_gray_mask, [vertices], 255)
    img_gray = cv2.bitwise_and(img_gray, img_gray_mask)

    img_color_mask = np.zeros_like(img)
    cv2.fillPoly(img_color_mask, [vertices], [255, 0, 0])
    img_color_masked = cv2.bitwise_and(img, img_color_mask)
    img_color_masked_intensity.append(img_color_masked.mean(axis=(0, 1, 2)))
    mask_max = np.ones_like(img)*255
    mask_max = cv2.bitwise_and(mask_max, img_color_mask)
    mask_max_intensity.append(mask_max.mean(axis=(0, 1, 2)))
```
