## Helper Functions

```python
def grayscale(img):
    """Applies the Grayscale transform
    Returns an image with only one color channel
    To display the returned image x as grayscale
    use plt.imshow(x, cmap='gray')
    """
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies the Gaussian noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    # define a blank mask to start with
    mask = np.zeros_like(img)
    
    # define a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape)>3:
        channel_count = img.shape[2] # can be 3 or 4 depending on the input image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    # filing pixels inside the polygon defined by `vertices` with the fill color
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    # return the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image

def draw_lines(img, lines, color=[255, 0, 0], thickness=2):
    """
    Draws `lines` with `color` and `thickness`.
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about
    combining this function with the weighted_img() function below.
    NOTE: `lines` is a list, and each `line` in `lines` 
    is a list as well
    """
    for line in lines:
        try:
            for x1, y1, x2, y2 in line:
                cv2.line(img, (x1, y1), (x2, y2), color, thickness)
        except:
            print(line)
            
def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap,
               color_in=[255, 0, 0], thickness_in=5):
    """
    `img` should be the output of a Canny transform
    
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), 
                            minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines, color=color_in, thickness=thickness_in)
    return line_img, lines
            
def weighted_img(img, initial_img, alpha=0.8, beta=1.0, gamma=0.0):
    """
    `img` is the output of the hough_lines(), an image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing
    
    The result image is computed as follows:
    
    initial_img * alpha + img * beta + gamma
    NOTE: `initial_img` and `img` must be the same shape
    """
    return cv2.addWeighted(initial_img, alpha, img, beta, gamma)

def fit_lanes(lines):
    """
    Fit lines for lanes
    """
    lane_left = []
    lane_right = []
    for line in lines:
        for x1, y1, x2, y2 in line:
            if (y2-y1)/(x2-x1)>0:
                lane_right.append([x1, y1])
                lane_right.append([x2, y2])
            else:
                lane_left.append([x1, y1])
                lane_left.append([x2, y2])

    points_left = np.squeeze(np.array(lane_left))
    points_right = np.squeeze(np.array(lane_right))

    y_min = min(min(points_left[:, 1]), min(points_right[:, 1]))
    y_max = max(max(points_left[:, 1]), max(points_right[:, 1]))
    y_min = np.int(y_max*0.3 + y_min*0.7)
        

    fit_left = np.polyfit(points_left[:, 0], points_left[:, 1], 1)
    fit_right = np.polyfit(points_right[:, 0], points_right[:, 1], 1)

    x_left_min = np.int((y_min-fit_left[1])/fit_left[0])
    x_left_min = min(959, x_left_min)
    x_left_max = np.int((y_max-fit_left[1])/fit_left[0])
    x_left_max = min(959, x_left_max)
    x_left = [x_left_min, x_left_max]
    
    y_left = np.polyval(fit_left, x_left).astype(np.int)
    x_right_min = np.int((y_min-fit_right[1])/fit_right[0])
    x_right_min = min(959, x_right_min)
    x_right_max = np.int((y_max-fit_right[1])/fit_right[0])
    x_right_max = min(959, x_right_max)
    x_right = [x_right_min, x_right_max]
    y_right = np.polyval(fit_right, x_right).astype(np.int)

    # plt.plot(x_left, np.polyval(fit_left, x_left))
    # plt.plot(x_right, np.polyval(fit_right, x_right))
    return [[[x_left[0], y_left[0], x_left[1], y_left[1]]], [[x_right[0], y_right[0], x_right[1], y_right[1]]]]
```

## Write the pipeline as a function

```python
def detect_lane(img):
    """
    Detects the lane lines in the image
    """
    
    # convert RGB to gray scale
    img_gray = grayscale(img)
    
    # Gaussian blur
    kernel_size = 5
    img_gray = gaussian_blur(img_gray, kernel_size)
    
    # Canny
    low_threshold = 30
    high_threshold = 90
    img_gray = canny(img_gray, low_threshold, high_threshold)
        
    # Find proper mask vertices
    left_bottom = [200, 539]
    right_bottom = [920, 539]
    apex = [520, 350]
        
    # Create mask
    # NOTE: cv2.fillPoly expects a list of np.array as pts, not a np.array!
    vertices = [np.array([left_bottom, right_bottom, apex])]
    img_masked = region_of_interest(img_gray, vertices)
    
    # Hough transform
    rho = 2
    theta = 2*np.pi/180
    hough_threshold = 25
    min_line_length = 25
    max_line_gap = 1
    _, lines = hough_lines(img_masked, rho, theta, hough_threshold, 
                           min_line_length, max_line_gap, thickness_in=10)
    
    # Fit lines
    try:
        lane_lines = fit_lanes(lines)
        img_line_fitted = np.zeros_like(img)
        draw_lines(img_line_fitted, lane_lines, thickness=10)
        img_superposed = cv2.addWeighted(img, 0.8, img_line_fitted, 1.2, 0)
    except:
        img_superposed = img
        
    return img_superposed
```
